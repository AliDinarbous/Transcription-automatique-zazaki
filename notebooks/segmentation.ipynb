```python
import os
import torch
import librosa
import soundfile as sf
from silero_vad import get_speech_timestamps, load_silero_vad
```

```python
input_audio = "/info/raid-etu/m1/s2506992/projet-m1-asr/datasets/audio_zazaki/Mealê Kirdî   Cizba Çaresine 14 30 (128kbit_AAC).m4a"
output_dir = "/info/raid-etu/m1/s2506992/projet-m1-asr/datasets/segments"
```

```python
os.makedirs(output_dir, exist_ok=True)
```

```python
model = load_silero_vad()
```

```python
wav, sr = librosa.load(input_audio, sr=16000)
```

```python
wav = torch.tensor(wav)
```

```python
speech_timestamps = get_speech_timestamps(
    wav,
    model,
    sampling_rate=16000,
    min_speech_duration_ms=1000,
    max_speech_duration_s=15,
    threshold=0.5
)
print(f"Nombre de segments détectés: {len(speech_timestamps)}")
```

```python
for i, segment in enumerate(speech_timestamps):
    start, end = segment['start'], segment['end']
    segment_audio = wav[start:end].cpu()

    sf.write(f"{output_dir}/segment_{i}.wav", segment_audio.numpy(), 16000)
```
