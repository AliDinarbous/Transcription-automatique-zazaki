```python
def add_noise(wav, snr_db):
    noise = torch.randn_like(wav)

    signal_power = wav.pow(2).mean()
    noise_power = noise.pow(2).mean()

    snr = 10 ** (snr_db / 10)
    scale = torch.sqrt(signal_power / (snr * noise_power))

    return wav + scale * noise
```

```python
# tester différents SNR
for snr in [5, 10, 15]:
    noisy = add_noise(wav, snr)
    print(f"Noisy SNR={snr}")
    display(Audio(noisy.numpy(), rate=sr))
```

```python
def add_noise_2(wav, snr_db):
    # normalisation
    wav = wav / (wav.abs().max() + 1e-8)

    noise = torch.randn_like(wav)

    signal_power = wav.pow(2).mean()
    noise_power = noise.pow(2).mean()

    snr = 10 ** (snr_db / 10)
    scale = torch.sqrt(signal_power / (snr * noise_power))

    noisy = wav + scale * noise

    return noisy
```
